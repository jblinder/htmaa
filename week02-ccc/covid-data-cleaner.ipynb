```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('public/data/owid-covid-data.csv')
df = df[df['location'] == 'United States']
df['new_date'] = pd.to_datetime(df['date'])

dfv = pd.read_csv('public/data/vaccinations/vaccinations.csv')
dfv = dfv[dfv['location'] == 'United States']
dfv['new_date'] = pd.to_datetime(dfv['date'])
```

```python
di = df.groupby(pd.Grouper(key='new_date', freq='1M')).sum() # groupby each 1 month
di.index = di.index.strftime('%B')

dv = dfv.groupby(pd.Grouper(key='new_date', freq='1M')).max() # groupby each 1 month
dv.index = dv.index.strftime('%B')
```
